## 7.9 Mortgage Calculator

```python
import numpy as np
import matplotlib.pyplot as plt

# Declare the variables
principal = 500_000
interest_rate = 0.05
years = 30
payments_year = 12
periods = years * payments_year
```

```python
# Create a function to calculate the monthly payment.
def calculate_monthly_payment(principal, interest_rate, years, payments_year):
    r = interest_rate / payments_year
    n = payments_year * years
    monthly_payment = principal * ((r * (1 + r) ** n) / ((1 + r) ** n - 1))
    return monthly_payment
```

```python
# Calculate the monthly payment
monthly_rate = interest_rate / payments_year
monthly_payment = calculate_monthly_payment(principal, interest_rate, years, payments_year)

# Print the monthly payment
print(f"The monthly payment is: {monthly_payment:0.2f}")
```

```python
# Calculate the principal and interest payments for each month.
# Create an array of the periods
def calculate_interest_and_principal_payments(principal, monthly_rate, monthly_payment, periods):
    interest_payments = np.zeros(periods)
    principal_payments = np.zeros(periods)

    for i in range(1, periods + 1):
        interest_payments[i - 1] = principal * monthly_rate
        principal_payments[i - 1] = monthly_payment - interest_payments[i - 1]
        principal = principal - principal_payments[i - 1]
    return interest_payments, principal_payments
```

```python
# Calculate the interest and principal payments
interest_payments, principal_payments = calculate_interest_and_principal_payments(
    principal, monthly_rate, monthly_payment, periods)

# Create a graph of the interest and principal payments
plt.plot(interest_payments, label="Interest Payments")
plt.plot(principal_payments, label="Principal Payments")
plt.legend()
plt.show()
```

```python
# Calculate the cumulative interest and principal payments
cumulative_interest_payments = np.cumsum(interest_payments)
cumulative_principal_payments = np.cumsum(principal_payments)

# Create a graph of the cumulative interest and principal payments
plt.plot(cumulative_interest_payments, label="Cumulative Interest Payments")
plt.plot(cumulative_principal_payments, label="Cumulative Principal Payments")
plt.legend()
plt.show()
```

## 7.11 Single Value Decomposition of an image (SVD)

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = Image.open("explorer.jpg").convert("L")
image = np.array(image)
```

```python
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
image.shape
```

```python
U, S, Vt = np.linalg.svd(image, full_matrices=False)

U.shape, S.shape, Vt.shape
```

```python
S = np.diag(S)

S.shape
```

```python
k = 4

component_k = np.outer(U[:, k], Vt[k, :]) * S[k,k]

plt.imshow(component_k, cmap="gray")
plt.axis("off")
plt.show()
```

```python
k = 100

compressed = U[:, :k] @ S[:k, :k] @ Vt[:k, :]

compressed.shape
```

```python
plt.imshow(compressed, cmap="gray")
plt.axis("off")
plt.show()
```

## 7.13 Simple linear regression

```python
n = 35

# Generate a random dataset for years of experience vs. salary
np.random.seed(0)
years_experience = np.random.rand(n) * 10  # Years of experience
salary = 30000 + 2000 * years_experience + np.random.randn(n) * 1000  # Salary with some noise
```

```python
# Plot the dataset without the regression line
plt.scatter(years_experience, salary)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Years of Experience vs. Salary (Data Only)")
plt.grid(True)
plt.show()
```

```python
# Calculate the coefficients of the linear regression model (y = mx + b)
x_mean = np.mean(years_experience)
y_mean = np.mean(salary)
xy_mean = np.mean(years_experience * salary)
x_squared_mean = np.mean(years_experience ** 2)

m = (xy_mean - x_mean * y_mean) / (x_squared_mean - x_mean ** 2)
b = y_mean - m * x_mean

# Calculate the predicted salaries
predicted_salary = m * years_experience + b
```

```python
# Plot the dataset and regression line
plt.scatter(years_experience, salary, label="Data Points")
plt.plot(years_experience, predicted_salary, color='red', label="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Years of Experience vs. Salary")
plt.legend()
plt.grid(True)
plt.show()
```

## 7.14 Ballistic trajectory

```python
g = 9.81

initial_velocity = np.array([30.0, 40.0, 50.0])
launch_angle = np.radians([30.0, 45.0, 60.0])

time = np.linspace(0, 6, 100)

fig, ax = plt.subplots(figsize=(8, 4))

for v0, angle in zip(initial_velocity, launch_angle):
    x = v0 * np.cos(angle) * time
    y = v0 * np.sin(angle) * time - 0.5 * g * time**2
    ax.plot(x, y)

ax.set_title("Cannonball Trajectories")
ax.set_xlabel("Horizontal Distance (m)")
ax.set_ylabel("Vertical Distance (m)")
ax.legend([f"Cannonball {i+1}" for i in range(len(initial_velocity))])
ax.set_xlim(0, np.max(x))
ax.set_ylim(0, np.max(y))
ax.grid()

plt.show()
```

## 7.15 Game of life

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Set the size of the grid and the number of generations
grid_size = 100
num_generations = 1000

# Initialize the grid with random values (0 or 1). On has a 20% chance

# grid = np.random.choice([0, 1], (grid_size, grid_size))

grid = np.zeros((grid_size, grid_size))

live_cells = [
    (50,49),(50,51),(51, 49),(51,51),(52,49),(52,50),(52,51),
    (20,21),(21,21),(22,21),(21,20),(20,22),(80,81),(81,81),
    (82,81),(81,80),(80,82),(20,81),(21,81),(22,81),(21,80),
    (20,82),(80,21),(81,21),(82,21),(81,20),(80,22)]

for row, col in live_cells:
    grid[row, col] = 1
```

```python
plt.imshow(grid, cmap='gray')
plt.axis("off")
```

```python
def get_neighbors(row, col):
    neighbors = 0
    for i in range(-1, 2):
        for j in range(-1, 2):
            if i == 0 and j == 0:
                continue
            if 0 <= row + i < grid_size and 0 <= col + j < grid_size:
                neighbors += grid[row + i, col + j]
    return neighbors
```

```python
def update_grid(grid):
    new_grid = np.copy(grid)
    for row in range(grid_size):
        for col in range(grid_size):
            neighbors = get_neighbors(row, col)
            is_alive = grid[row, col] == 1
            if is_alive:
                if not 2 <= neighbors <= 3:
                    new_grid[row, col] = 0
            else:
                if neighbors == 3:
                    new_grid[row, col] = 1
    return new_grid
```

```python
# Create a function to animate the grid
def animate(i):
    # Mat here is the matplotlib object that we can manipulate
    mat.set_data(update_grid(grid))
    grid[:] = mat.get_array()
    return [mat]
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
plt.axis("off")
mat = ax.matshow(grid, cmap='gray')
anim = animation.FuncAnimation(fig, animate, interval=50, frames=num_generations, blit=True)
```
