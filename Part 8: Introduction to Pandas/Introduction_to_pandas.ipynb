```python
#@title Generate a random .csv file

import csv
import random

# Define the number of rows and columns
num_rows = 5
num_columns = 3

# Generate random data and write it to a CSV file
with open('random_data.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)

    # Write the header row
    header = ['Column1', 'Column2', 'Column3']
    csv_writer.writerow(header)

    # Generate and write random data for each row
    for _ in range(num_rows):
        row = [random.randint(1, 100) for _ in range(num_columns)]
        csv_writer.writerow(row)

print("Random CSV file generated as 'random_data.csv'")
```

## 8.4 Time series data

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('flights')

df.head(10)
```

```python
df.set_index(['year', 'month'], inplace=True)

df.head()
```

```python
df.reset_index(inplace=True)

df.head()
```

```python
# 1949-Jan
df['date'] = df['year'].astype(str) + '-' + df['month'].astype(str)

df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])

df.head()
```

```python
df.set_index('date', inplace=True)

df.head()
```

```python
passengers = df.passengers

passengers.head()
```

```python
df.reset_index(inplace=True)

df.drop(columns=['date'], inplace=True)

df.head()
```

```python
df = df.pivot(index='year', columns='month', values='passengers')

df.head()
```

## 8.5 Visualizing our data

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset('penguins')
flights = sns.load_dataset('flights')
```

### 8.5.1 Penguins dataset visualization

Which species has a highest average flipper length?

```python
flipper_length_means = penguins.groupby('species')['flipper_length_mm'].mean()

flipper_length_means.plot(kind='line', marker="o")
```

Which island has the largest penguin population?

```python
populations = penguins['island'].value_counts()

populations.plot(kind="bar")
```

What is the relationship between bill length and flipper length?

```python
penguins.plot.scatter(x="bill_length_mm", y="flipper_length_mm")
```

```python
penguins['bill_length_mm'].plot.hist(bins=20)
```

```python
penguins.boxplot(column='flipper_length_mm', by="species")
```

### 8.5.2 Flights dataset visualization

```python
pivot_flights = flights.pivot(index="month", columns="year", values="passengers")

pivot_flights.head()
```

```python
total_passengers = flights.groupby("year")["passengers"].sum()

total_passengers.plot(kind="line")
```

```python
average_passengers_month = flights.groupby("month")["passengers"].sum()

average_passengers_month.plot(kind="bar")
```

```python
pivot_flights.plot()
```
